### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x_line = np.linspace(0, 10, 100)
y_line = x_line ** 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot the line chart in the first subplot
axs[0, 0].plot(x_line, y_line)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x values')
axs[0, 0].set_ylabel('y values')

# Plot the bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Chart')

# Plot the scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter)
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f"({x_scatter[i]}, {y_scatter[i]})", (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(0, 10), ha='center')
axs[1, 0].set_title('Scatter Plot')

# Plot the pie chart in the fourth subplot
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add a main title for the entire subplot layout
fig.suptitle('2x2 Subplot Layout')

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()
plt.show()
```
